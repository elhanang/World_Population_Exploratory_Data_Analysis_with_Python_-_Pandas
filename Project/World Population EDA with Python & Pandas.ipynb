```python
df.corr(numeric_only = True)

# This compares every column to every other column and looks at how closely correalated they are.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (20, 7)

sns.heatmap(df.corr(numeric_only=True), annot=True)


plt.show()


# This imports the matplotlib and seaborn libraries and visualizes the data with a heatmap. This shows how correlated the data is to one another.
```

```python
df.groupby('Continent').mean(numeric_only=True).sort_values(by="2022 Population",ascending=False)

# This groups the data by the values in the "Continent" column, calculates the mean (average) of all numeric columns for each continent group, and sorts the resulting table based on the average 2022 population per continent, from highest to lowest.
```

```python
df[df['Continent'].str.contains('Oceania')]

# This filters the DataFrame df to return only the rows where the Continent column contains the word 'Oceania'.
```

```python
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only=True).sort_values(by="2022 Population",ascending=False)
df2

# Groups the data by continent.
# Selects only the columns related to population across various years.
# Computes the average population per continent for each of those years.
# Sorts the resulting table by average 2022 population, from highest to lowest.
# Stores the result in a new DataFrame called df2.
# Displays the final grouped, averaged, and sorted DataFrame.
```

```python
df.columns

# It returns the column labels (names) of the DataFrame df.
```

```python
df3 = df2.transpose()
df3

# It transposes the DataFrame df2 — that means it swaps the rows and columns.
```

```python
df3.plot()

# It creates a line plot from the transposed DataFrame df3.
```

```python
df.boxplot(figsize=(20,10))

# It creates a box plot for all numeric columns in your DataFrame df.
# The parameter figsize=(20,10) just sets the size of the plot to make it big and readable.
```
